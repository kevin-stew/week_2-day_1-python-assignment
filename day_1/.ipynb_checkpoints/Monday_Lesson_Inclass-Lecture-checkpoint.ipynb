#### Performing Calculations on ints and floats

##### Addition

```python
num1 = 2
num2 = 5.4

result = num1 + num2

print(result)

# short-hand version
result += 2 # result = result + 2
print(result)
```

##### Subtraction

```python
result_diff = num2 - num1
print(result_diff)

# Short-hand
result_diff -= 1
print(result_diff)
```

##### Multiplication

```python
result_mul = num2 * num1
print(result_mul)

# Short-hand
reulst_mul *= 2
```

##### Division

```python
result_quo = num2 / num1

print(result_quo)

#shorthand
result_quo /= 2
print(result_quo)
```

##### Floor Division

```python
result_floor = num2 // num1
print(result_floor)

#shorthand
result_floor //= 2
print(result_floor)
```

##### Modulo

```python
result_mod = num2 % num1
print(result_mod)

#shorthand
result_mod %= 1
print(result_mod)
```

```python
square = 5 ** 2
print(square)

#shorhand
square **= 2
print(square)
```

### If Statements

```python
# Available operators: Greater(>), Less(<),Equal(==)
# Greater or Equal(>=), Less or Equal (<=)

# Truth Tree:
# T && F = F
# T && T = T
# T || F = T
# F || T = T
# F || F = F

num1 = 10
num2 = 10

if num2 == num1:
    print("Equal Values")
else:
    print("Not Equal")
    
if num2 < num1:
    print("num2 is less")
elif num1 <= num2:
    print("Num1 is less")    
elif num2 == num1:
    print("equal valuse")
else:
    print("what kind of numbers are these?!")
    
```

## Using 'and'/'or' with If Statements

```python
num_11 = 15
num_12 = 3
num_13 = 10
num_14 = 3

# If with "and f statement
if num_11 / 5 == num_12 and num_13 - 7 == num_14:
    print("Both are True")
    
# If with "or" statement
if num_11 > num_12 or num_13 == num_14:
    print("At least one of theses is true")
```

### Elif Statements

```python
first_name = "Max"

if first_name == "Smith":
    print("The name is Smith.")
elif first_name != "Max":
    print("The name is NOT Max.")
else:
    print("The name is Max.")
```

# Exercise 3 <br>
<p>Take in a users input for their age, if they are younger than 18 print kids, if they're 18 to 65 print adults, else print seniors</p>

```python
age = int(input("Please enter your age: "))

if age < 18:
    print("You are just a kid.  Go have some fun and don't worry about the future!")
elif 18 <= age <= 65:
    print("You are an Adult.  I hope you like working!")
else:
    print("You are a Senior. Enjoy your Social Security checks!")
```
